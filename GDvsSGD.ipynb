```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
x= pd.read_csv('features.csv')
y= pd.read_csv('labels.csv')
# Conversion en tableaux numpy
X = x.values
y = y.values
w = np.zeros((6,))
b = 0

def cost(X, y, w, b):
    m = len(y)
    total_cost = 0.0
    for i in range(m):
        f_wb_i = np.dot(X[i], w) + b
        total_cost += (f_wb_i - y[i])**2
    cost = total_cost / (2*m)
    return cost

def compute_gradient(X, y, w, b): 
    m,n = X.shape          
    dj_dw = np.zeros((n,))
    dj_db = 0.

    for i in range(m):                             
        err = (np.dot(X[i], w) + b) - y[i]   
        for j in range(n):                         
            dj_dw[j] = dj_dw[j] + err * X[i, j]    
        dj_db = dj_db + err                        
    dj_dw = dj_dw / m                                
    dj_db = dj_db / m                                
        
    return dj_db, dj_dw

# Descente de gradient
alpha = 0.000000005
tolerance = 1e-6
max_iter = 1000000
costs = []
iteration = 0
while iteration < max_iter:
    dj_db, dj_dw = compute_gradient(X, y, w, b)
    w = w - alpha * dj_dw
    b = b - alpha * dj_db
    cost_i = cost(X, y, w, b)
    costs.append(cost_i)
    if len(costs) > 1 and np.abs(costs[-1] - costs[-2]) < tolerance:
        break
    iteration += 1

# Affichage des résultats
print("Nombre d'itérations:", iteration)
print("Coût final:", cost_i)
print("Poids optimisés:", w)
print("Biais optimisé:", b)
#test pour la 1 ere ligne du dataset price = 8565.44
# Produit de la première ligne de X avec w
w = np.array(w)
prediction = np.dot(X[0], w)

print("Prediction aevc vals 1ere ligne :",prediction)
```

```python
import random
x= pd.read_csv('features.csv')
y= pd.read_csv('labels.csv')
# Conversion en tableaux numpy
X = x.values
y = y.values
w = np.zeros((6,))
b = 0

def cost(X, y, w, b):
    m = len(y)
    total_cost = 0.0
    for i in range(m):
        f_wb_i = np.dot(X[i], w) + b
        total_cost += (f_wb_i - y[i])**2
    cost = total_cost / (2*m)
    return cost

def compute_gradient_stochastique(X, y, w, b, i): 
    dj_dw = np.zeros((len(w),))
    dj_db = 0.
    err = (np.dot(X[i], w) + b) - y[i]   
    for j in range(len(w)):                         
        dj_dw[j] = err * X[i, j]    
    dj_db = err                        
        
    return dj_db, dj_dw

# Descente de gradient stochastique
alpha =  0.000000005
tolerance = 1e-6
max_iter = 1000000
costs = []
iteration = 0
while iteration < max_iter:
    shuffled_indices = list(range(len(y)))
    random.shuffle(shuffled_indices)
    for i in shuffled_indices:
        dj_db, dj_dw = compute_gradient_stochastique(X, y, w, b, i)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
    cost_i = cost(X, y, w, b)
    costs.append(cost_i)
    if len(costs) > 1 and np.abs(costs[-1] - costs[-2]) < tolerance:
        break
    iteration += 1

# Affichage des résultats
print("Nombre d'itérations:", iteration)
print("Coût final:", cost_i)
print("Poids optimisés:", w)
print("Biais optimisé:", b)

#test pour la 1 ere ligne du dataset price = 8565.44
# Produit de la première ligne de X avec w
w = np.array(w)
prediction = np.dot(X[0], w)

print("Prediction aevc vals 1ere ligne :",prediction)
```

La GD calcule les gradients de toutes les données d'entraînement pour chaque mise à jour de poids et de biais, tandis que la SGD calcule les gradients pour un exemple de données à la fois. La SGD est plus rapide que la GD pour les grands ensembles de données, elle permet d'obtenir une convergence plus rapide et une solution plus robuste.
